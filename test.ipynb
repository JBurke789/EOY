```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#read file and make lists
file = '2022.csv'
data = pd.read_csv(file)

dates = data['Date'].tolist()
artists = data['Artist '].tolist()
genres = data['Genre'].tolist()
years = data['Year'].tolist()
```

## Genre Pie Chart

```python
def gen_genre_analysis(genre):
    #remove leading and trailing white space so all genre names the same
    clean_genres=[]
    for i in genre:
        clean_genres.append(i.strip())
    labels = set(clean_genres)#create list of unique genres present
    #count instances of unique genres
    counts = []
    for i in labels:
        counts.append(clean_genres.count(i))

    fig1,ax1 = plt.subplots()
    explode = np.zeros_like(np.array(counts)) + 0.05
    patches, texts, autotexts =ax1.pie(counts,
        labels=labels,
        autopct='%1.1f%%',
        shadow=True,
        explode=explode,
        startangle=90,
        pctdistance=0.85,
        labeldistance=1.1)

    #change text colour
    for text in texts:
        text.set_color('grey')
    for autotext in autotexts:
        autotext.set_color('black')

    #make sircle
    centre_circle = plt.Circle((0,0),0.70,fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)


    ax1.axis('equal')
    plt.tight_layout()
    plt.show()

    
```

```python
gen_genre_analysis(genres)
```
